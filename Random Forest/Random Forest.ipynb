```python
#Import Libraries
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=1000
sns.set_theme(style='darkgrid',palette='rainbow')
import warnings
warnings.filterwarnings('ignore')
```

# Regression

```python
#Read Data
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/snehuuu28/Data_Science/main/Random%20Forest/Boston.csv',index_col=0)
```

```python
df
```

```python
df.isnull().sum()
```

# Model Building

```python
x = df.iloc[:,:-1]
y = df['medv']
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
#Evaluting Regressor
```

```python
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)
```

```python
print('R2 value:',r2_score(ytest,ypred))
```

```python
print(f"R2 values: {r2_score(ytest, ypred)}")       # f is Format String
print(f"MSE:{mean_squared_error(ytest,ypred)}")
print(f"RMSE:{np.sqrt(mean_squared_error(ytest,ypred))}")
```

```python
#R2 value is 0 to 1
# Higher the R2 value then model is good

#RMSE: 0 to Infinity

# Therefore our model is Good
```

# Classification

```python
#Read the Dataset
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/snehuuu28/Data_Science/main/Random%20Forest/Mobile_price_range.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x= df['price_range'], palette='Set2')
```

```python
df['price_range'].replace([0,1,2,3],['low','medium','high','very high'], inplace = True)
```

```python
df
```

```python
#Train Model
```

```python
x = df.iloc[:,:-1]
y = df['price_range']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain,ytrain)
yp = rfc.predict(xtest)
```

```python
yp
```

```python
print(classification_report(ytest,yp)) 
```

```python
len(rfc.estimators_)
```

```python
list(x.columns)
```

```python
list(df['price_range'].unique())
```

```python
tree.plot_tree(rfc.estimators_[0], feature_names =list(x.columns),class_names=list(df['price_range'].unique()), filled=True, max_depth=3)
plt.show()
```
