# 2) Write a program to read a year from user , find whether that year is leap or not

```python
year = int(input("Enter the Year:"))
if year % 4 == 0:
    if year % 100 != 0:
        if year % 400 == 0:
            print("Leap")
        else:
            print("Not Leap")
    else:
        print("Leap")
else:
    print("Not Leap")
```
