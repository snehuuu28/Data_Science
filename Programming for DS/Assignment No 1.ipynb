```python
R = float(input("Enter the Radius : "))
Area = 3.142*R*R
print("Area = ", Area)

P = 2*3.14*R
print("Perimeter = ",P)
```

```python
# The amount & percentage of interest from the keyboard & find final
```

```python
Amount = int(input('Enter Amount : '))
Per = int(input("Enter the Percentage of interest : "))
interest = Amount * (Per/100)
Final = Amount + interest
print("Final Amount : ",Final)
```

```python
# Distance value in meter & convert into centimeter,inches & Yard
```

```python
Meter = float(input("Enter the value : "))
Centimeter = Meter*100
print("Centimeter : ",Centimeter)
Inches = Meter * 39.3701
print("Inches : ",Inches)
Yard = Meter * 1.09361
print("Yard : ",Yard)
```
