# Factorial of Number

```python
num=int(input("Enter the no:"))
fact = 1
while (num>0):
    fact=fact*num
    num=num -1
print("Factorial=",fact)
```

```python
num=int(input("Enter the no:"))
fact = 1
counter=1
while (counter<=num):
    fact=fact*counter
    counter=counter+1
print("Factorial=",fact)
```

# Sum of Digits of this number



```python
num = int(input("Enter any num:"))
sum = 0
reminder=0
while(num > 0):
    reminder = num % 10
    sum = sum + reminder
    num = num//10
print(sum)
```

# Write a Program to read number from user and print its Reverse

```python
num = int(input("Enter any num:"))
reverse = 0
digit=0
while(num > 0):
    reverse = num % 10
    digit = (digit*10)+ reverse
    num=num//10
print("reverse=",digit)
    
    
```

# Write a Program to Print Fibonacci Series up to given Number
