```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame({
    'Marks': [85,87,85,88,900],
    'Time' : [5,6,6.6,5,5.5]
})
df
```

```python
iso = IsolationForest()
iso.fit(df)
outlier = iso.predict(df)
```

```python
outlier
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/snehuuu28/Machine_Learning/main/EDA-2/Wholesale%20customers%20data.csv")
```

```python
df
```

```python
sns.pairplot(df)
```

```python
df = df[['Delicassen','Frozen']]
df
```

```python
df.boxplot()
```

```python
pe.box(df['Delicassen'], orientation='h')
```

```python
iso = IsolationForest()
iso.fit(df)
outliers = iso.predict(df)
outliers
```

```python
df['Outliers'] = outliers
df
```

```python
pe.box(df['Frozen'], orientation='h')
```

```python
df[df['Outliers']==-1]
```
