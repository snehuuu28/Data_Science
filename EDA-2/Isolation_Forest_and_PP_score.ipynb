```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.ensemble import IsolationForest
%matplotlib inline
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Dummy dataset
df = pd.DataFrame({
    'Marks' : [85,89,87,88,800],
    'Time' : [1,1.2,1.5,2,20]
})
```

```python
df
```

```python
outlier = iso.predict(df)

outlier
```
